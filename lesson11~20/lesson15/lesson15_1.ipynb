```python
import numpy as np
import matplotlib.pyplot as plt
index=np.array([2,3,4,5,6])
subject=['apple','banana','tomato','orange','guava']
value1=[2,3,4,5,6]
value2=[4,5,6,7,8]
value3=[1,2,3,4,5]

bin_width=0.3

figure , axes = plt.subplots()
axes.bar(index,value1,width=bin_width,color='yellow',label='value1')
axes.bar(index+bin_width,value2,width=bin_width,color='red',label='value2')
axes.bar(index-bin_width,value3,width=bin_width,color='green',label='value3')
axes.legend(loc='upper center',ncols=3)
axes.set_xticks(index,labels=subject)

xmin,xmax,ymin,ymax=axes.axis()

xmin-=1
xmax+=1
ymin+=1
ymax+=1
axes.axis((xmin,xmax,ymin,ymax))

plt.show()
```
